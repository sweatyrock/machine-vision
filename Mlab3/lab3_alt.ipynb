```python
#Task: binarize image
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
import matplotlib.image as pltimg

fig = plt.figure(figsize=(15,10))
imlena = pltimg.imread('.\\lena_cover.jpg')
imlena = np.mean(imlena,-1)

tt = np.max(imlena)
imlena_bin = imlena > 70

plt.subplot(221)
plt.imshow(imlena, cmap=plt.cm.gray)
plt.title('Original image')
plt.subplot(222)
plt.imshow(imlena_bin, cmap=plt.cm.gray)
plt.title('Result')
plt.show()
```

```python

imcam = data.camera()
imcam_bin = imcam > 30
fig, axes = plt.subplots(ncols=2, figsize=(15, 10))
ax = axes.ravel()

ax[0].imshow(imcam, cmap=plt.cm.gray)
ax[0].set_title('Original image')

ax[1].imshow(imcam_bin, cmap=plt.cm.gray)
ax[1].set_title('Result')

for a in ax:
    a.axis('off')

plt.show()
```

```python
#task: apply skeletonize and erosion to binarized images
from skimage.morphology import disk
from skimage.morphology import erosion, skeletonize
fig = plt.figure(figsize=(15,10))
selem = disk(6)
imlena_bin_eroded = erosion(imlena_bin, selem)
imlena_bin_skel = skeletonize(imlena_bin == 0)
imcam_bin_eroded = erosion(imcam_bin, selem)
imcam_bin_skel = skeletonize(imcam_bin == 0)
plt.subplot(321)
plt.title('binary')
plt.imshow(imlena_bin, cmap=plt.cm.gray)
plt.subplot(322)
plt.title('eroded')
plt.imshow(imlena_bin_eroded, cmap=plt.cm.gray)
plt.subplot(323)
plt.title('skeletonized')
plt.imshow(imlena_bin_skel, cmap=plt.cm.gray)
plt.subplot(324)
plt.title('binary')
plt.imshow(imcam_bin, cmap=plt.cm.gray)
plt.subplot(325)
plt.title('eroded')
plt.imshow(imcam_bin_eroded, cmap=plt.cm.gray)
plt.subplot(326)
plt.title('skeletonized')
plt.imshow(imcam_bin_skel, cmap=plt.cm.gray)
plt.show()
```

```python
#apply canny to images
from skimage import feature
fig = plt.figure(figsize=(15,10))
imlena_canny = feature.canny(imlena,sigma=3)
imcam_canny = feature.canny(imcam,sigma=3)
plt.subplot(221)
plt.title('original')
plt.imshow(imlena, cmap='gray')
plt.subplot(222)
plt.title('canny, sigma = 3')
plt.imshow(imlena_canny, cmap='gray')
plt.subplot(223)
plt.title('original')
plt.imshow(imcam, cmap = 'gray')
plt.subplot(224)
plt.title('canny, sigma = 3')
plt.imshow(imcam_canny, cmap = 'gray')
plt.show()
```
